## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_data = pd.merge(mouse_metadata, study_results, on='Mouse ID')

# # Display the data table for preview
mouse_data
```

```python
# Checking the number of mice.
mouse_data['Mouse ID'].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
md_duplicates = mouse_data[mouse_data.duplicated(['Mouse ID','Timepoint'])]
md_duplicates['Mouse ID']
```

```python
# Get all the data for the duplicate mouse ID. 
mouse_data.loc[mouse_data['Mouse ID'] == 'g989']
```

```python
# clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = mouse_data.loc[mouse_data['Mouse ID'] != 'g989']
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_df['Mouse ID'].nunique()
```

## Summary Statistics

```python
# summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

drug_group = clean_df.groupby(["Drug Regimen"])

drug_group_mean = drug_group['Tumor Volume (mm3)'].mean()
drug_group_median = drug_group['Tumor Volume (mm3)'].median()
drug_group_var = drug_group['Tumor Volume (mm3)'].var()
drug_group_std = drug_group['Tumor Volume (mm3)'].std()
drug_group_SEM = drug_group['Tumor Volume (mm3)'].sem()

drug_group_dict = {
    "Mean": drug_group_mean,
    "Median":drug_group_median,
    "Variance":drug_group_var,
    "STD":drug_group_std,
    "SEM":drug_group_SEM,
}

drug_summary_df = pd.DataFrame(drug_group_dict)
drug_summary_df
```

```python
# Using the aggregation method

drug_group['Tumor Volume (mm3)'].agg(['mean','median','var','std','sem'])
```

## Bar and Pie Charts

## Quartiles, Outliers and Boxplots

```python
# final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
mouse_group = clean_df.groupby(['Mouse ID'])
mouse_max = mouse_group['Timepoint'].max()
mouse_max_df = pd.DataFrame(mouse_max)

mouse_merge = pd.merge(clean_df, mouse_max_df, on='Mouse ID')
tumor_vol_loc = mouse_merge.loc[(mouse_merge['Timepoint_x'] == mouse_merge['Timepoint_y'])]
tumor_vol_df = tumor_vol_loc[['Mouse ID','Tumor Volume (mm3)','Drug Regimen']]
tumor_vol_df
```

```python
#finding any outliers in each of the 4 most succesful regimen
treatment_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

tumor_vol_list = []

for x in treatment_list:    
    
    filtered = tumor_vol_df.loc[tumor_vol_df['Drug Regimen'] == x]['Tumor Volume (mm3)']
    tumor_vol_list.append(filtered)
    
    quartiles = filtered.quantile([.25,.5,.75])
    
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    outlier = filtered.loc[(filtered > upper_bound) | (filtered < lower_bound)]
    
    print(f'Outliers for {x} are: {outlier}')
```

```python
green_diamond = dict(markerfacecolor='g', marker='D')

fig1, ax1 = plt.subplots()
ax1.set_ylabel('Tumor Size')
ax1.boxplot(tumor_vol_list, labels=treatment_list, flierprops=green_diamond)
plt.show()
```

## Line and Scatter Plots

```python
# line plot of tumor volume vs. time point for a randomly selected mouse treated with Capomulin
cap_df = clean_df.loc[clean_df['Drug Regimen'] == 'Capomulin']

mouse = cap_df['Mouse ID'].sample().values[0]

rando_mouse = cap_df[cap_df['Mouse ID'] == mouse]

mouse_x = rando_mouse['Timepoint']

mouse_y = rando_mouse['Tumor Volume (mm3)']

mouse_plot = plt.plot(mouse_x, mouse_y, color='green', marker='.', label=mouse)

plt.title("Tumor Size Over Time")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")

handles = mouse

plt.legend(handles=mouse_plot, loc="best")
# maximum is tumor size for Capomulin 48.15820939
plt.xlim(0,45)
plt.ylim(20,50)

plt.show()
```

```python
# scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

cap_id_group = cap_df.groupby(['Mouse ID'])
weight = cap_id_group['Weight (g)'].mean()
av_tum_size = cap_id_group['Tumor Volume (mm3)'].mean()

plt.scatter(av_tum_size, weight, marker="o", facecolors="blue", edgecolors="black", alpha=0.90)

plt.ylim(weight.min()-1, weight.max()+1)
plt.xlim(av_tum_size.min()-1, av_tum_size.max()+1)

plt.title("Average Tumor Volume Vs. Mouse Weight")
plt.xlabel("Average Tumor Volume")
plt.ylabel("Weight (g)")
```

## Correlation and Regression

```python
x_values = av_tum_size
y_values = weight

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values, y_values, marker="o", facecolors="blue", edgecolors="black", alpha=0.90)
plt.plot(x_values,regress_values,"r-")
plt.title("Average Tumor Volume Vs. Mouse Weight")
# plt.annotate(line_eq,(2,20),fontsize=10,color="red")
plt.xlabel("Average Tumor Volume")
plt.ylabel("Weight (g)")
plt.show()

correlation = st.pearsonr(y_values,x_values)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```
